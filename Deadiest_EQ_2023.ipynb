```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("/content/Deadliest earthquakes by year.csv")
```

```python
df1=df1[[#'Unnamed: 0', 
 'Year', 'Magnitude', 'Location', 'Depth (km)', 
 #'MMI',
       #'Notes', 
 'Event', 'Date']].copy()
```

```python
df1
```

```python
df1['Location'].value_counts()
```

```python
df1['Location'].str.split(",")
```

```python
df1['Location'].str.split(",").str.get(0)
```

```python
Country=df1['Location'].str.split(",").str.get(0)
```

```python
df1["Country"] = Country
```

```python
df1
```

```python
EQ =pd.DataFrame(df1,columns=["Year","Date", "Magnitude","Depth (km)", "Country"])
```

```python
EQ.info()
```

```python
EQ['Magnitude']=pd.to_numeric(EQ['Magnitude'],errors = "coerce")
```

```python
EQ["Depth (km)"] = pd.to_numeric(EQ["Depth (km)"],errors = "coerce")
```

```python
EQ
```

```python
EQ.index
```

```python
EQ.set_index("Year")
```

```python
EQ.dtypes
```

```python
EQ.dropna(inplace = True)
```

```python
EQ
```

```python
EQ['Country'].value_counts()
```

```python
df_pivot = EQ.pivot(index="Year",columns='Country', values="Magnitude")
df_pivot.info()
```

```python
deadiest_eq=df_pivot[["Iran","China","Turkey","Indonesia","Soviet Union","India","Japan"]]
```

```python
deadiest_eq.plot(kind = "line" , figsize=(8,4), xlabel=" Year" , ylabel="Mag" ,title = "lineplot showing Magnitude of deadiest EQ ")
plt.show()
```

```python
deadiest_eq.index
```

```python
df_pivot_2023=deadiest_eq[deadiest_eq.index.isin([2023])]
```

```python
df_pivot_2023
```

```python
df_pivot_2023=df_pivot_2023.T
```

```python
df_pivot_2023.plot(kind = "barh" , figsize=(8,4), xlabel=" Year" , ylabel="Country" ,title = "Barplot showing Magnitude of deadiest EQ in 2023 " )

plt.show()
plt.savefig("deadiest_2023.png")
```

```python
df_pivot_years=deadiest_eq[deadiest_eq.index.isin([ 1986, 1987, 1988, 1989, 1990,1990 ,2018, 2019, 2020, 2021, 2022, 2023])]
```

```python
df_pivot_years.plot(kind = "bar"  ,  figsize=(8,4), xlabel=" Year" , ylabel="Mag" ,title = "Barplot showing Magnitude of deadiest EQ " , width = 5)
plt.show()
plt.savefig("bar_2023.png")
```
