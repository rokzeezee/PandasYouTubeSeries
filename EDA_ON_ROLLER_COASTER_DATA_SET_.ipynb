TO GET THE AVERGAE SPEED AND AREA WIH MORE COASTERS

```python
ax= df.query("Location != 'other'").groupby("Location")["speed_mph"] .agg(["mean" ,"count"]).query("count>= 10").sort_values("mean").plot(kind = "barh" , figsize = (12,5) , title = "Average coast speed and location")
ax.set_xlabel("Average coaster speeed")
plt.show()
```
