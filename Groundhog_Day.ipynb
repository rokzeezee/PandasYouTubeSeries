```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
resp =requests.get('https://groundhog-day.com/api/v1/groundhogs')
resp.json()
```

```python
df_groundhod=pd.DataFrame(resp.json()['groundhogs'])
df_groundhod.columns
```

```python
df_groundhod=df_groundhod[['id', 'slug', 'shortname', 'name', 'city', 'region', 'country',
       'coordinates', #'source',
  'contact', #'currentPrediction',
  'isGroundhog',
       'type', 'active', 'successor', 'description', # 'image',
       #'predictionsCount', #'predictions'
  ]].copy()
```

```python
df_groundhod.head()
```

```python
df_groundhod['lat']=df_groundhod['coordinates'].str.split(',').str[0]
df_groundhod['long'] =df_groundhod['coordinates'].str.split(',').str[1]
```

```python
df_groundhod['lat']=df_groundhod['lat'].astype(float)
df_groundhod['long']=df_groundhod['long'].astype(float)
```

```python
df_groundhod.drop('coordinates' , axis = 1 , inplace = True)
```

```python
df_groundhod.dtypes
```

# New section

```python
import folium
```

```python
map = folium.Map(location = [df_groundhod['lat'].mean() , df_groundhod['long'].mean()], zoom_start = 4)

for _, row in df_groundhod.iterrows():
  folium.Marker(location=[row['lat'] , row['long']] , popup= row['slug']).add_to(map)
map
```
